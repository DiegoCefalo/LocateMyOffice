- Queremos tener restaurantes veganos cerca porque el jefe es vegano y si lo mantengo contento me da un ascenso. (40% de peso)

- Hay que tomar encuenta los  Starbucks porque sin café los ejecutivos no sirven. (10% de peso)

- Finalmente, al conserje le gusta el basket y es muy majo, asi que le toca un estadio cerca. (10% de peso)

```python
sr.puntajilizador('LA',los_angeles)
```
